### Strength-Depth Profile
In class I introduced the strengh-depth profile as a way of summarizing the strength of the lithosphere. 
The next exercise will be to plot one of these diagrams. 

We need three things to plot a simple strength depth profile:
1. An equation defining the brittle strength at shallow depth. For this we will use a Byerlee's law:

$\tau = 0.85\sigma_n$ when $\sigma_n < 200$ MPa<br>
$\tau = 60 + 0.6\sigma_n$, when $\sigma_n > 200$ MPa<br>
where $\sigma_n = \sigma_{lith}$

2. An equation defining the temperature as a function of depth. This is needed to calculate the viscous strength. We will use a profile called the half-space cooling model which described the conductive cooling of the lithosphere as a function of its age: 

$T = T_s + (T_m - T_s)erf(y/(2\sqrt{\kappa t}))$

3. An equation defining the viscous strength of the lithosphere.

$\eta = A_o\exp(E/(R T))$

Recall that the viscous strength is

$\sigma = 2 \eta \dot{\epsilon}$

The strength is then defined as the minimum of the brittle and viscous strength at each depth.
